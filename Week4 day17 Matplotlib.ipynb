# Q1.Create a line plot and scatter plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Category': ['A', 'B', 'C', 'D'] * 3,  
    'Values': [45, 30, 15, 10] * 3,       
    'X': np.arange(1, 13),              
    'Y': np.random.randint(10, 100, 12)
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(7, 5))
plt.plot(df['X'], df['Y'], marker='o', color='b', label='Line Plot')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.scatter(df['X'], df['Y'], color='r', label='Scatter Plot')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Q2. Create bar chat and pie chart

```python
plt.plot(df['X'], df['Y'], marker='o', color='b', label='Line Plot')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
```

```python
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
