```python
# Q1. What function in Seaborn is commonly used to create a scatter plot, and how can you add a regression line to it?
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    'x': range(10),
    'y': [2 * i + 1 + (i % 2) * 3 for i in range(10)]
})

sns.regplot(x='x', y='y', data=data)
plt.title("Scatter Plot with Regression Line")
plt.show()
```

```python
# Q2. How can you customize the appearance of Seaborn plots using themes?
```

```python
sns.histplot(data['y'], kde=True, color='green', bins=6)
plt.title("Histogram with KDE Overlay")
plt.show()
```

```python
# Q3. Which Seaborn functions are best suited for visualizing categorical data?
```

```python
data1 = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Value': [5, 7, 8, 6, 9, 5]
})

sns.barplot(x='Category', y='Value', data=data1)
plt.title("Bar Plot")
plt.show()
```

```python
# Q4. How do you create a heatmap in Seaborn to visualize a correlation matrix?
```

```python
import numpy as np

heatmap_data = pd.DataFrame({
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10),
    'D': np.random.rand(10)
})

corr_matrix = heatmap_data.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
